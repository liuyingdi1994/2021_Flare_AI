```python
data_path = './datasets/12/data.csv'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(data_path)
data.head()
```

```python
data = np.array(data)
X, y = data[:, :2], data[:, 2]
X.shape
y.shape
```

```python
from collections import Counter
Counter(y)
```

```python
y_0 = plt.scatter(X[y==0, 0], X[y==0, 1])
y_1 = plt.scatter(X[y==1, 0], X[y==1, 1])
plt.legend((y_0, y_1), ('y = 0', 'y = 1'))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2021)
X_train.shape
y_train.shape
X_test.shape
y_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
mlp = Sequential()
```

```python
mlp.add(Dense(units=20, input_dim=2, activation='sigmoid'))
mlp.add(Dense(units=1, activation='sigmoid'))
```

```python
mlp.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
mlp.summary()
```

```python
mlp.fit(X_train, y_train, batch_size=128, epochs=5000, verbose=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_predict = mlp.predict_classes(X_test)
accuracy_score(y_true=y_test, y_pred=y_predict)
```

```python
# 绘制决策边界
def plot_decision_boundary(model, axis, epsilon=500):
    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], epsilon).reshape(-1, 1),
        np.linspace(axis[2], axis[3], epsilon).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D'])
    
    plt.contourf(x0, x1, zz, cmap=custom_cmap)
```

```python
plot_decision_boundary(model=mlp, axis=[-0.1, 1.1, -0.1, 1.1], epsilon=500)
y_0 = plt.scatter(X[y==0, 0], X[y==0, 1])
y_1 = plt.scatter(X[y==1, 0], X[y==1, 1])
plt.legend((y_0, y_1), ('y = 0', 'y = 1'))
plt.show()
```
